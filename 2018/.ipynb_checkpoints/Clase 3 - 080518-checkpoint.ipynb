### Imprimimos en la pantalla el incremento de una variable entera

```java
int posX = 0;
String miNuevoTexto = "";
String miTexto = "Cantidad: ";
void setup(){
    size(400, 400);
    textSize(40);
}
void draw(){
    background(0);
    posX = posX + 1;
    miNuevoTexto = miTexto + posX;
    text(miNuevoTexto,0,200);
}
```

### Mostramos un texto descendiendo con el incremento de una variable entera

```java
int posX = 0;
String miNuevoTexto = "";
String miTexto = "Cantidad: ";
void setup(){
    size(400, 400);
    textSize(40);
}
void draw(){
    background(0);
    posX = posX + 1;
    miNuevoTexto = miTexto + posX;
    text(miNuevoTexto,posX,posX);
}
```

# DivisiÃ³n (/) y MÃ³dulo (%)
Devuelve el resto de una divisiÃ³n. Por ejemplo:

10/3 = 3,3333

10%3 = 1

40%10 = 0

5%2 = 1

```java
int inc, division, modulo;
String miNuevoTexto;
void setup(){
    size(400, 400);
    textSize(30);
    inc = 0;
    modulo = 0;
}
void draw(){
    background(0);
    inc = inc + 1;
    division = floor(inc / 40);
    modulo = inc % 40;
    miNuevoTexto = "Division: " + inc + " / 40 = " + division;
    text(miNuevoTexto,10,100);
    miNuevoTexto = "MÃ³dulo: " + inc + " % 40 = " + modulo;
    text(miNuevoTexto,10,200);
}
```

## Random 
Devuelve un nÃºmero aleatorio en a y b
random(10) -> Un valor entre 0 y 10 con decimales (float)
random(5, 9) -> Un valor entre 5 y 9 con decimales (float)

## round 
Redondea el valor entero 
7.3 -> 7
7.8 -> 8

## floor
Me devuelve el entero
7.3 -> 7
7.8 -> 7

## ceil
Me devuelve el siguiente valor entero
7.3 -> 8
7.8 -> 8

```java
float valorFloat, valorInt;
String miNuevoTexto;
void setup(){
    size(400, 400);
    textSize(30);
    valorFloat = random(0, 10);
    valorInt = round(valorFloat);
    valorInt = floor(valorFloat);
    valorInt = ceil(valorFloat);
}
void draw(){
    background(0);
    miNuevoTexto = "Random: " + valorFloat;
    text(miNuevoTexto,0,100);
    miNuevoTexto = "Random: " + valorInt;
    text(miNuevoTexto,0,300);
}
```
